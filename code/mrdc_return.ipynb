#### Data 

```python
# load the .csv file for current dataset
def load_dataset(month):
    df = pd.read_csv(path1 + "monthly/"+ month + ".csv")
    df.columns = df.columns.str.lower()
    return df
 
```

```python
# filter the returned checks and divide data into bins 
returned_check = pd.DataFrame()
def returned_checks(df, month): 
    returned_df = df[df['status'] == 'returned']
    returned_check[month] = returned_df.groupby(pd.cut(returned_df['score'], bins=range(0,801,100)), observed=False).size()
    return returned_check
```

##### Returned checks with mrdc score < 800 

```python
# count 

months = [ 'sept_23', 'oct_23', 'nov_23', 'dec_23', 'jan_24']
for i in range(len(months)): 
    df = load_dataset(months[i])
    returned_checks(df, months[i])
returned_check
```

```python
# percentage 

total_counts = returned_check.sum(axis=0)
returned_check_percentage = returned_check.copy()
for month in returned_check_percentage.columns:
    returned_check_percentage[month] = round((returned_check_percentage[month] / total_counts[month]) * 100, 2)
    
returned_check_percentage
```

##### Returned Amount 

```python
return_amount = pd.DataFrame()
def returned_amount(df, month): 
    returned_df = df[df['status'] == 'returned']
    return_amount[month] = returned_df['amount'].groupby(pd.cut(returned_df['score'], bins=range(0,801,100)), observed=False).mean()
    return return_amount

```

```python
# monthly returned amount 
for i in range(len(months)): 
    df = load_dataset(months[i])
    returned_amount(df, months[i])
return_amount
```

#### Combine Reports 

```python
from bs4 import BeautifulSoup
import os
import datetime
from dateutil.relativedelta import relativedelta

path1 = "/Users/pooja/Desktop/Pooja/mrdc/data/"
path2 = "/Users/pooja/Desktop/Pooja/mrdc/reports/"

def combine_html(files, headings, main_heading, output_file):
  with open(output_file, "w") as output:
    output.write("<!DOCTYPE html>\n<html>\n<head>\n")
    output.write('<meta charset="UTF-8">')
    output.write('<title>Combined Report</title>')

    # CSS for centered and highlighted headings
    output.write("""
      <style>
        .highlight {
          text-align: center;
          background-color: #ddd;
          padding: 10px;
          border: 1px solid #ddd;
        }
        .heading {
            text-align: center;
            padding: 10px;
        }
      </style>
    """)
    output.write(f"<h1 class='heading'>{main_heading}</h1>")
    output.write("\n</head>\n<body>\n")

    # Combine contents with styled headings
    for file, heading in zip(files, headings):
      with open(file, "r") as f:
          content = f.read()
          soup = BeautifulSoup(content, "html.parser")

          output.write(f"<h2 class='highlight'>{heading}</h2>")
          output.write(str(soup))
      output.write("\n\n")

    output.write("\n</body>\n</html>")

   
```

```python
months = ['sept_23', 'oct_23', 'nov_23', 'dec_23', 'jan_24']

for month in months: 
    file1 = path2 + "mrdc_model_drift/model_performance_" + month + ".html"
    file2 = path2 + "mrdc_data_drift/data_drift_" + month+ ".html"
    file3 = path2 + "mrdc_data_drift/data_stability_"+month+ ".html"

    files = [file1, file2, file3]
    heading1 = "Model Performance"
    heading2 = "Dataset Drift"
    heading3 = "Dataset Stability"
    main_heading = month+" Report"
    output_file = path2 + 'combined/report_' +month+'.html'  
    headings = [heading1, heading2, heading3]

    combine_html(files, headings, main_heading, output_file)
```
